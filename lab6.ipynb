```python

import boto3
from botocore.exceptions import ClientError
import json

def get_secret(secret_name):
    region_name = "us-east-1"

    # Create a Secrets Manager client
    session = boto3.session.Session()
    client = session.client(
        service_name='secretsmanager',
        region_name=region_name
    )

    try:
        get_secret_value_response = client.get_secret_value(
            SecretId=secret_name
        )
    except ClientError as e:
        raise e

    secret = get_secret_value_response['SecretString']
    
    return json.loads(secret)
```

```python
import matplotlib.pyplot as plt

# Calculate year over year income growth rate for each state
df['income_growth_rate'] = df.groupby('name')['income'].pct_change()

# Pivot the dataframe to get years as columns and states as rows for the income growth rate
df_pivot_growth = df.pivot(index='name', columns='year', values='income_growth_rate')

# Plotting
plt.figure(figsize=(15,10))
for state in df_pivot_growth.index:
    plt.plot(df_pivot_growth.columns, df_pivot_growth.loc[state], marker='', linewidth=1, label=state)

plt.legend(loc='best')
plt.title('Year Over Year Income Growth Rate by State')
plt.xlabel('Year')
plt.ylabel('Income Growth Rate')
plt.grid()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Calculate income percent change by state
df['income_percent_change'] = df.groupby('name')['income'].pct_change()

# Pivot the dataframe to get years as columns and states as rows
df_pivot = df.pivot(index='name', columns='year', values='income_percent_change')

# Plotting
plt.figure(figsize=(15,10))
for state in df_pivot.index:
    plt.plot(df_pivot.columns, df_pivot.loc[state], marker='', linewidth=1, label=state)

plt.legend(loc='best')
plt.title('Income Percent Change by State over Time')
plt.xlabel('Year')
plt.ylabel('Income Percent Change')
plt.grid()
plt.show()
```
